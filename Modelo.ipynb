3. Construa um modelo de regressão linear múltipla utilizando as variáveis independentes (AT, V, AP, RH)
para prever a variável dependente (PE). Divida os dados em treino (80%) e teste (20%).

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
entrada = dados.drop(columns=["PE"])
saida = dados["PE"]
```

```python
scaler  = MinMaxScaler(feature_range=(0, 1))
padronizados  = scaler.fit_transform(dados)
dados_padronizados = pd.DataFrame(padronizados, columns=dados.columns)
dados_padronizados
```

```python
entrada_treino, entrada_teste, saida_treino, saida_teste = train_test_split(entrada, saida, test_size=0.3, random_state=42)

print(f"Tamanho do dataset = {dados.shape[0]} amostras")
print(f"Quantidade de amostras de treinamento = {entrada_treino.shape[0]} amostras")
print(f"Quantidade de amostras de teste = {entrada_teste.shape[0]} amostras")
```

```python
modelo = LinearRegression()
modelo.fit(entrada_treino, saida_treino)
```

4. Avalie o desempenho do modelo utilizando métricas como R², RMSE e MAE. O modelo é adequado para
prever a produção de energia? Justifique sua resposta.

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
saida_pred = modelo.predict(entrada_teste)
```

```python
mse = mean_squared_error(saida_teste, saida_pred)
r2 = r2_score(saida_teste, saida_pred)
mae = mean_absolute_error(saida_teste, saida_pred)


print(f"Erro quadrático médio (MSE): {mse:.2f}")
print(f"Coeficiente de determinação (R²): {r2:.2f}")
print(f"Erro médio absoluto (MAE): {mae:.2f}")
```
